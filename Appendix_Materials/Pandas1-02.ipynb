# Filtering DAtaFrames with many Conditions (AND)

```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head(10)
```

```python
mask1 = titanic.sex == "male"
mask1.head()
```

```python
mask2 = titanic.age > 14
mask2.head()
```

```python
(mask1 & mask2).head()
```

```python
male_surv = titanic.loc[mask1 & mask2, ["survived", "pclass", "sex", "age"]]
male_surv.head(20)
```

```python
male_surv.info()
```

```python
male_surv.describe()
```

```python
titanic.describe()
```

# Filtering DataFrames with many Conditions (OR)

# apply(), map(), and applymap()

```python
import pandas as pd
```

```python
sales = pd.read_csv("sales.csv", index_col = 0)
```

```python
sales
```

```python
sales.info()
```

```python
sales.min(axis = 0)
```

```python
sales.min(axis = 1)
```

```python
def range(series):
    return series.max() - series.min()
```

```python
sales.apply(range, axis = 0)
```

```python
sales.apply(range, axis = 1)
```

```python
sales.apply(lambda series: series.max() - series.min(), axis=1)
```

```python
sales.apply(lambda x: x.max() - x.min(), axis=1)
```

```python
summer = pd.read_csv("summer.csv")
```

```python
summer.head()
```

```python
summer.Athlete.apply(lambda x: x[0])
```

```python
summer.Athlete.map(lambda x: x[0])
```

```python
summer.iloc[:,1:3].applymap(lambda x: x[0])
```

```python
sales.applymap(lambda x: 0.4*x-5)
```
